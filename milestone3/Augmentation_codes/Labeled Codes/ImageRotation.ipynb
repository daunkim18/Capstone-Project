```python
import os

# Define label directory
label_dir = os.path.join(os.getcwd(), "scripts", "data", "labels", "train")
label_filenames = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

# Save rotated label files here
rotated_label_dir = os.path.join(r"C:\Users\Reshmi\Capstone-Project\milestone3\Augmented_images", "rotated_txt")
os.makedirs(rotated_label_dir, exist_ok=True)

# Rotate bounding boxes 90° clockwise
def rotate_yolo_bbox_90(x, y, w, h):
    new_x = y
    new_y = 1.0 - x
    return new_x, new_y, w, h

# Process each label file
for filename in label_filenames:
    label_path = os.path.join(label_dir, filename)
    rotated_label_path = os.path.join(rotated_label_dir, f"rotated_{filename}")

    with open(label_path, 'r') as infile, open(rotated_label_path, 'w') as outfile:
        for line in infile:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            class_id, x_center, y_center, width, height = parts
            x_center, y_center, width, height = map(float, (x_center, y_center, width, height))
            new_x, new_y, new_w, new_h = rotate_yolo_bbox_90(x_center, y_center, width, height)
            outfile.write(f"{class_id} {new_x:.6f} {new_y:.6f} {new_w:.6f} {new_h:.6f}\n")

    print(f"Saved rotated labels: {rotated_label_path}")

print("Rotation (90°) of YOLO labels complete!")
```
