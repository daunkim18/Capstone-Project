```python
import cv2
import os
import matplotlib.pyplot as plt

# Define image directory
image_dir = folder_path = os.path.join(os.getcwd(), "../Resources/Dental_Images/Dental_Images/")

# Define list of image filenames
image_filenames = ["2.jpg", "3.jpg", "4.jpg", "5.jpg", "7.jpg", "9.jpg", "12.jpg", "13.jpg", "14.jpg", 
                   "17.jpg", "16.jpg", "17.jpg", "18.jpg", "19.jpg", "20.jpg", "22.jpg", "23.jpg", 
                   "24.jpg", "25.jpg", "26.jpg", "27.jpg", "30.jpg"]

# Create a directory to save resized images
save_dir = os.path.join(image_dir, "resized_images")
os.makedirs(save_dir, exist_ok=True)  # Create folder if it doesn't exist

# Define new size
new_width = 250
new_height = 250
new_size = (new_width, new_height)

# Process each image
for filename in image_filenames:
    image_path = os.path.join(image_dir, filename)

    # Load the image
    img = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if img is None:
        print(f"Error: Failed to load {filename}")
        continue  # Skip to next image

    # Resize the image
    rescaled_img = cv2.resize(img, new_size, interpolation=cv2.INTER_LINEAR)

    # Save the resized image
    save_path = os.path.join(save_dir, f"resized_{filename}")
    cv2.imwrite(save_path, rescaled_img)
    print(f"Saved resized image: {save_path}")

print("âœ… Image resizing complete!")

# Load a saved image to verify
test_image = cv2.imread(os.path.join(save_dir, "resized_2.jpg"))
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for matplotlib

# Display the image
plt.imshow(test_image)
plt.axis("off")  # Hide axis
plt.show()

```

```python
import cv2
import os
import matplotlib.pyplot as plt

# Define directories
image_dir = folder_path = os.path.join(os.getcwd(), "../Resources/Dental_Images/Dental_Images/")
save_dir = os.path.join(image_dir, "cropped_images")
os.makedirs(save_dir, exist_ok=True)  # Create folder if it doesn't exist

# List of image filenames
image_filenames = ["image2.jpg", "82.jpg", "41.jpg", "73.jpg", "62.jpg", "51.jpg", "49.jpg", "30.jpg", "22.jpg", "26.jpg", "7.jpg", "53.jpg", "65.jpg", "57.jpg", "71.jpg", "69.jpg", "14.jpg", "31.jpg", "45.jpg", "17.jpg", "39.jpg"]

# **Fixed cropping area (modify these values as needed)**
y1, y2 = 50, 150  # Height range (rows)
x1, x2 = 50, 200  # Width range (columns)

# Process each image
for filename in image_filenames:
    image_path = os.path.join(image_dir, filename)

    # Load the image
    img = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if img is None:
        print(f"Error: Failed to load {filename}")
        continue  # Skip to next image

    print(f"Original image shape for {filename}: {img.shape}")

    # Ensure cropping dimensions are within the image size
    h, w = img.shape[:2]  # Get height and width of the image
    if y2 > h or x2 > w:
        print(f"Skipping {filename}: Cropping area exceeds image dimensions")
        continue

    # **Fixed Cropping**
    cropped_image = img[y1:y2, x1:x2]

    print(f"Cropped image shape for {filename}: {cropped_image.shape}")

    # Save the cropped image
    save_path = os.path.join(save_dir, f"cropped_{filename}")
    cv2.imwrite(save_path, cropped_image)
    print(f"Saved cropped image: {save_path}")

    # Display cropped image
    cropped_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
    plt.imshow(cropped_rgb)
    plt.axis("off")
    plt.title(f"Cropped Image - {filename}")
    plt.show()

print("Cropping process complete!")
```
