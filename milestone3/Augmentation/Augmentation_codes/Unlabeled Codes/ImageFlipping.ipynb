```python
import cv2
import os
import matplotlib.pyplot as plt

# Define original image directory
image_dir = r'C:\Users\Reshmi\Capstone-Project\milestone3\Raw_images'

# Define output directory for flipped images
save_dir = r'C:\Users\Reshmi\Capstone-Project\milestone3\Augmented_images\Unlabeled\flipped_images'
os.makedirs(save_dir, exist_ok=True)

# Get up to 1000 .jpg files
image_filenames = [
    fname for fname in os.listdir(image_dir)
    if fname.lower().endswith(".jpg")
][:1000]

# Flip and save images
for fname in image_filenames:
    image_path = os.path.join(image_dir, fname)
    img = cv2.imread(image_path)

    if img is None:
        print(f"Error: Failed to load {fname}")
        continue

    # Flip horizontally (flipCode=1)
    flipped_img = cv2.flip(img, 1)

    save_path = os.path.join(save_dir, f"flipped_{fname}")
    cv2.imwrite(save_path, flipped_img)
    print(f"Saved: {save_path}")

print("All images flipped and saved in a separate folder!")

# --- Display the first flipped image ---

flipped_files = [f for f in os.listdir(save_dir) if f.lower().endswith(".jpg")]
if flipped_files:
    first_image_path = os.path.join(save_dir, flipped_files[0])
    test_image = cv2.imread(first_image_path)
    test_image_rgb = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)

    plt.imshow(test_image_rgb)
    plt.axis("off")
    plt.title(f"Sample: {flipped_files[0]}")
    plt.show()
else:
    print("No flipped images found to display.")
```
