```python
import os

# Define label directory
label_dir = os.path.join(os.getcwd(), "scripts", "data", "labels", "train")
label_filenames = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

# Save flipped label files here
flipped_label_dir = os.path.join(r"C:\Users\Reshmi\Capstone-Project\milestone3\Augmented_images", "flipped_txt")
os.makedirs(flipped_label_dir, exist_ok=True)

# Flip bounding boxes horizontally
for filename in label_filenames:
    label_path = os.path.join(label_dir, filename)
    flipped_label_path = os.path.join(flipped_label_dir, f"flipped_{filename}")

    with open(label_path, 'r') as infile, open(flipped_label_path, 'w') as outfile:
        for line in infile:
            parts = line.strip().split()
            if len(parts) != 5:
                continue  # Skip malformed lines
            class_id, x_center, y_center, width, height = parts
            # Flip x_center (horizontal flip)
            x_center_flipped = 1.0 - float(x_center)
            # Write flipped line
            outfile.write(f"{class_id} {x_center_flipped:.6f} {y_center} {width} {height}\n")

    print(f"Saved flipped labels: {flipped_label_path}")

print("Label flipping complete!")
```
