```python
import os

# Input parameters
original_width = 640
original_height = 480
new_width = 416
new_height = 416

# Define label directory with absolute pixel values
label_dir = os.path.join(os.getcwd(), "scripts", "data", "labels", "train")
label_filenames = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

# Output directory for resized normalized YOLO labels
resized_label_dir = os.path.join(r"C:\Users\Reshmi\Capstone-Project\milestone3\Augmented_images", "resized_txt")
os.makedirs(resized_label_dir, exist_ok=True)

# Convert absolute to YOLO-normalized coordinates (after resize)
for filename in label_filenames:
    label_path = os.path.join(label_dir, filename)
    resized_label_path = os.path.join(resized_label_dir, f"resized_{filename}")

    with open(label_path, 'r') as infile, open(resized_label_path, 'w') as outfile:
        for line in infile:
            parts = line.strip().split()
            if len(parts) != 5:
                continue

            class_id, x, y, w, h = map(float, parts)

            # Resize x/y/w/h according to new dimensions
            x = x * new_width / original_width
            y = y * new_height / original_height
            w = w * new_width / original_width
            h = h * new_height / original_height

            # Normalize for YOLO
            x_norm = x / new_width
            y_norm = y / new_height
            w_norm = w / new_width
            h_norm = h / new_height

            outfile.write(f"{int(class_id)} {x_norm:.6f} {y_norm:.6f} {w_norm:.6f} {h_norm:.6f}\n")

    print(f"Saved resized label: {resized_label_path}")

print("Absolute to YOLO-normalized resizing complete!")
```
